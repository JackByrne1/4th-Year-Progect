```
import pandas as pd
import numpy as np
```

```
from google.colab import files
uploaded = files.upload()
import io
df = pd.read_csv(io.BytesIO(uploaded['ASC3TWzf9.csv']))
```

```
df.replace( to_replace=-9999.90,
    value=np.nan,
    inplace=True,
    limit=None,
    regex=False,
    method='pad',)
df.replace( to_replace=-9999.9004,
    value=np.nan,
    inplace=True,
    limit=None,
    regex=False,
    method='pad',)
df.replace( to_replace=-10000.0,
    value=np.nan,
    inplace=True,
    limit=None,
    regex=False,
    method='pad',)
final = df.drop(0).astype(float).interpolate()
final.tail(60)
```

```
.to_csv('data.csv', index=False)
files.download('data.csv')
```

```
final['DateTime'] = 1009843200 + final.day*86400 + final.hr*3600
final.tail()
```

```
final['DateTime'] = pd.to_datetime(final["DateTime"], unit='s')
final.tail(60)
```
